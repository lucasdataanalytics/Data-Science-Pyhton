# 1.  **Importação da Base de Dados**

```python
import pandas as pd
```

```python
# URL do DataSet

url = "https://raw.githubusercontent.com/lucasdataanalytics/Data-Science-Pyhton/main/CreditCardChurn/Dataset/ClientesBanco.csv"
```

```python
tabela = pd.read_csv(url, encoding="latin1")
tabela = tabela.drop("CLIENTNUM", axis=1)
display(tabela)
```

# 2. **Preparação e Limpeza do Dados**

```python
# Verificar e Excluir Valores Vazios

tabela = tabela.dropna()
display(tabela.info())
```

```python
# Análise Descritiva das Informações

display(tabela.describe().round(1))
```

# 3. **Análise Exploratória de Dados (EDA)**

## Divisão Entre Clientes X Cancelados

```python
qtd_categoria = tabela["Categoria"].value_counts()
display(qtd_categoria)
```

```python
# Verificar Percentual do Total

qtd_categoria_perc = tabela["Categoria"].value_counts(normalize=True)
display(qtd_categoria_perc)
```

# 4. Visualizações e Insights

## Gráficos

```python
# Importar biblioteca de Dataviz

import plotly.express as px
```

```python
# Contagem de Produtos por Categoria
fig_categoria = px.histogram(tabela, x='Produtos Contratados', color="Categoria", title='Produtos Contratados por Categoria')

# Distribuição da idade dos clientes
fig_idade = px.histogram(tabela, x='Idade', color="Categoria", nbins=30, title='Distribuição da Idade dos Clientes por Categoria')

# Relação entre Contato 12M e Categoria
fig_salario_limite = px.histogram(tabela, x='Contatos 12m', color='Categoria', title='Contatos por Categoria')

# Relação entre Limite Disponível e Categoria
fig_limite_utilizacao = px.histogram(tabela, x='Limite Disponível', color='Categoria',
                                   title='Relação entre Limite Disponível por Categoria')
# Relação entre Qtd de Transações e Categoria
fig_total_transacoes = px.histogram(tabela, x='Qtde Transacoes 12m', color='Categoria',
                                   title='Quantidade de Transações por Categoria')

# Exibir os gráficos
fig_categoria.show()
fig_idade.show()
fig_salario_limite.show()
fig_limite_utilizacao.show()
fig_total_transacoes.show()
```

## Insights
---


*   Clientes que fizeram entre 40-60 transações nos últimos 12 meses tem uma maior Taxa de Cancelamento
*   As chances de cancelamento aumentam quanto maior for a quantidade de contatos
*   Quanto maior for os produtos contratados, menor a chance de Cancelamento
